```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gender= data['Gender'].value_counts()
```

**Barchart**

```python
plt.figure(figsize=(10, 6))
gender.plot(kind='bar', color='orange')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
health_counts = data['Health Status'].value_counts()

plt.figure(figsize=(10, 6))
health_counts.plot(kind='bar', color='purple')
plt.title('Distribution of Health Status')
plt.xlabel('Health Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
state_counts = data['State'].value_counts()

plt.figure(figsize=(10, 6))
state_counts.plot(kind='bar', color='teal')
plt.title('Distribution of States')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Histogram**

```python
income_intervals = [0, 20000, 40000, 60000, 80000, 100000]

income = data['Income']

plt.figure(figsize=(10, 10))

plt.hist(income, bins=income_intervals, color='lightgreen', edgecolor='black')
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.7)
plt.xticks(income_intervals)
plt.show()

```

```python
class_intervals = [0, 20, 40, 60, 80]

ages = data['Age']

plt.figure(figsize=(10, 10))

plt.hist(ages, bins=class_intervals, color='blue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(class_intervals)
plt.show()
```
